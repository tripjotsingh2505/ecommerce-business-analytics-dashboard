```python
df['order_purchase_timestamp'] = pd.to_datetime(df['order_purchase_timestamp'])

monthly_orders = (
    df
    .groupby(df['order_purchase_timestamp'].dt.to_period('M'))
    .agg(order_count=('order_id', 'nunique'))
    .reset_index()
)

monthly_orders['order_purchase_timestamp'] = monthly_orders['order_purchase_timestamp'].astype(str)

plt.figure(figsize=(12,5))
sns.lineplot(data=monthly_orders, x='order_purchase_timestamp', y='order_count')
plt.xticks(rotation=45)
plt.title("Monthly Order Trend")
plt.xlabel("Month")
plt.ylabel("Number of Orders")
plt.show()

# Insight:
# - Orders show an overall increasing trend.
# - Certain months show spikes, indicating seasonal demand.
# - Business can plan inventory & marketing around peak months.
```

```python
# QUESTION 1: MONTHLY REVENUE TREND OVER TIME?
```

```python
# date column fix -->
df["order_purchase_timestamp"] = pd.to_datetime(df["order_purchase_timestamp"])
```

```python
# delivered orders only -->
delivered_df = df[df["order_status"] == "delivered"]
```

```python
# INSIGHTS -->
# Revenue shows a clear upward trend over time
# Initial months have low revenue (business early stage)
# Revenue stabilizes and peaks in later months
# Indicates business growth and customer adoption
```

```python
# CUSTOMER CONTRIBUTION TO TOTAL REVENUE?
```

```python
# total revenue per customer -->
customer_revenue = (
    df.groupby("customer_unique_id")["payment_value"]
    .sum()
    .reset_index()
)
```

```python
# sort customers by revenue in descening order -->
customer_revenue = customer_revenue.sort_values(
    by="payment_value", ascending=False
)
```

```python
# feature engineering (cumulative revenue calculation) -->
customer_revenue["cumulative_revenue"] = customer_revenue["payment_value"].cumsum()
total_revenue = customer_revenue["payment_value"].sum()

customer_revenue["cumulative_revenue_pct"] = (
    customer_revenue["cumulative_revenue"] / total_revenue * 100
)
```

```python
# cumulative customer percentage -->
customer_revenue["customer_rank"] = range(1, len(customer_revenue) + 1)
customer_revenue["customer_pct"] = (
    customer_revenue["customer_rank"] / len(customer_revenue) * 100
)
```

```python
# visualisation (pareto chart) -->
plt.figure()
plt.plot(customer_revenue["customer_pct"],
         customer_revenue["cumulative_revenue_pct"])
plt.axhline(80, linestyle='--')
plt.xlabel("Percentage of Customers")
plt.ylabel("Percentage of Total Revenue")
plt.title("Customer Contribution to Total Revenue (Pareto Analysis)")
plt.show()
```

```python
# INSIGHT EXTRACTION (how many customers generate 80% revenue?) -->
# A small percentage of customers contribute a large portion of total revenue, indicating a strong Pareto effect.
# Approximately X% of customers are responsible for nearly 80% of revenue.
# Retaining and engaging these customers is critical for sustained business growth.
```

```python
# REGION WISE DELIVERY PERFORMANCE?
```

```python
# convert to datetime
df['order_delivered_customer_date'] = pd.to_datetime(df['order_delivered_customer_date'])
df['order_estimated_delivery_date'] = pd.to_datetime(df['order_estimated_delivery_date'])
```

```python
# delivery status
df['delivery_status'] = df['order_delivered_customer_date'] <= df['order_estimated_delivery_date']
```

```python
# region-wise aggregation
region_perf = (
    df.groupby('customer_state')
    .agg(
        total_orders=('order_id', 'count'),
        on_time_orders=('delivery_status', 'sum')
    )
    .reset_index()
)

region_perf['delayed_orders'] = region_perf['total_orders'] - region_perf['on_time_orders']
region_perf['on_time_percentage'] = (
    region_perf['on_time_orders'] / region_perf['total_orders']
) * 100
```

```python
# sort by performance
region_perf = region_perf.sort_values('on_time_percentage', ascending=False)

region_perf.head()
```

```python
# visualisation (bar chart) -->
plt.figure(figsize=(12,6))
plt.bar(region_perf['customer_state'], region_perf['on_time_percentage'])
plt.xticks(rotation=90)
plt.xlabel("Region / State")
plt.ylabel("On-time Delivery Percentage")
plt.title("Region-wise On-time Delivery Performance")
plt.tight_layout()
plt.show()
```

```python
# INSIGHTS --> Key Insights:
# Certain regions consistently show higher on-time delivery rates, indicating better logistics infrastructure.
# Regions with low on-time percentages may suffer from:
   # Longer shipping distance
   # Fewer delivery partners
   # Poor last-mile connectivity
   # High-order-volume regions with low performance are critical risk areas for customer satisfaction.

# Business Recommendations:
# Improve logistics partnerships in low-performing regions.
# Introduce buffer delivery timelines for historically delayed regions.
# Offer region-based delivery promises to reduce negative reviews.
```

```python
# STATES WITH HIGHEST DELIVERY DELAYS?
```

```python
# convert dates
df['order_delivered_customer_date'] = pd.to_datetime(df['order_delivered_customer_date'])
df['order_estimated_delivery_date'] = pd.to_datetime(df['order_estimated_delivery_date'])
```

```python
# calculate delay days -- feature engineering (we are adding one column "delay days") -->
df['delay_days'] = (
    df['order_delivered_customer_date'] - df['order_estimated_delivery_date']
).dt.days
```

```python
# consider only delayed deliveries
delayed_df = df[df['delay_days'] > 0]
```

```python
# state-wise delay analysis
state_delay = (
    delayed_df.groupby('customer_state')
    .agg(
        avg_delay_days=('delay_days', 'mean'),
        delayed_orders=('order_id', 'count')
    )
    .reset_index()
)
```

```python
# top states with highest average delay
state_delay = state_delay.sort_values('avg_delay_days', ascending=False)

state_delay.head(10)
```

```python
# visualisation (bar plot) -->
top_states = state_delay.head(10)

plt.figure(figsize=(10,6))
plt.bar(top_states['customer_state'], top_states['avg_delay_days'])
plt.xlabel("State")
plt.ylabel("Average Delay (Days)")
plt.title("Top 10 States with Highest Delivery Delays")
plt.tight_layout()
plt.show()
```

```python
# INSIGHT --> Key Insights:
# Some states experience significantly higher average delivery delays, indicating regional logistics challenges.
# High delay duration often correlates with distance from sellers and warehouse hubs.
# States with both high delay days and high delayed order count are critical problem zones.

# Business Recommendations:
# Optimise warehouse placement closer to high-delay regions.
# Improve last-mile delivery partnerships in these states.
# Add proactive customer communication for deliveries in delay-prone regions.
```

```python
# average review score: delayed vs on-time delivery -->
delay_review = (
    df.groupby("delayed")["review_score"]
    .mean()
    .reset_index()
)

delay_review["delivery_status"] = delay_review["delayed"].map({
    False: "On-time Delivery",
    True: "Delayed Delivery"
})

delay_review
```

```python
# visualisation: impact on customer satisfaction -->
plt.figure(figsize=(6,4))
plt.bar(
    delay_review["delivery_status"],
    delay_review["review_score"]
)
plt.ylabel("Average Review Score")
plt.title("Impact of Delivery Delay on Customer Satisfaction")
plt.ylim(0,5)
plt.show()
```

```python
# Optional (STRONG INSIGHT): Review Distribution -->
plt.figure(figsize=(6,4))
df.boxplot(column="review_score", by="delayed")
plt.title("Review Score Distribution: On-time vs Delayed")
plt.suptitle("")
plt.xlabel("Delayed Delivery")
plt.ylabel("Review Score")
plt.show()
```

```python
# INSIGHTS --> Findings:
# Delayed orders have significantly lower average review scores
# On-time deliveries ≈ 4+ rating
# Delayed deliveries drop close to 3 or below

# Customer Behavior:
# Customers punish delays more than product issues
# Even good products get low ratings if delivery is late
```
