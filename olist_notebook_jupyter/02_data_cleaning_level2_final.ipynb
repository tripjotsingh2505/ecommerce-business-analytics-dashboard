```python
# LEVEL-2 CLEANING  AND MERGING
```

```python
import pandas as pd
import numpy as np
```

```python
orders = pd.read_csv("clean_olist_orders.csv", parse_dates=[
    "order_purchase_timestamp",
    "order_approved_at",
    "order_delivered_carrier_date",
    "order_delivered_customer_date",
    "order_estimated_delivery_date"
])

customers = pd.read_csv("clean_olist_customers.csv")
order_items = pd.read_csv("clean_olist_order_items.csv")
payments = pd.read_csv("clean_olist_order_payments.csv")
products = pd.read_csv("clean_olist_products.csv")
reviews = pd.read_csv("clean_olist_order_reviews.csv")
sellers = pd.read_csv("clean_olist_sellers.csv")
category_translation = pd.read_csv("clean_product_category_name_translation.csv")
```

```python
print(orders.shape)
print(customers.shape)
print(order_items.shape)
print(payments.shape)
print(products.shape)
print(reviews.shape)
print(sellers.shape)
print(category_translation.shape)
```

```python
# orders table + customers table (base table = orders).
orders_customers = orders.merge(
    customers,
    on="customer_id",
    how="left"
)
```

```python
# orders table + payments table (base table = orders).
orders_payments = orders_customers.merge(
    payments,
    on="order_id",
    how="left"
)
```

```python
# orders table + order_items table (base table = orders).
orders_items = orders_payments.merge(
    order_items,
    on="order_id",
    how="left"
)
```

```python
# orders table + order_products table (base table = orders).
orders_products = orders_items.merge(
    products,
    on="product_id",
    how="left"
)
```

```python
# adding product_category table.
orders_products = orders_products.merge(
    category_translation,
    on="product_category_name",
    how="left"
)
```

```python
# adding order_reviews table.
orders_reviews = orders_products.merge(
    reviews,
    on="order_id",
    how="left"
)
```

```python
# adding order_sellers table (which is a last table) and now, we'll make final dataset table -->
final_df = orders_reviews.merge(
    sellers,
    on="seller_id",
    how="left"
)
```

```python
# LEVEL-2 OFFICIAL CLEANING AFTER JOINING/ MERGIING TABLES.
```

```python
# removing logical inconsistencies.
final_df = final_df[
    final_df["order_delivered_customer_date"]
    >= final_df["order_purchase_timestamp"]
]
```

```python
# handle NULLs created by joins.
final_df["review_score"] = final_df["review_score"].fillna(0)
final_df["payment_value"] = final_df["payment_value"].fillna(0)
```

```python
# feature engineering.
final_df["delivery_days"] = (
    final_df["order_delivered_customer_date"]
    - final_df["order_purchase_timestamp"]
).dt.days

final_df["delay_flag"] = (
    final_df["order_delivered_customer_date"]
    > final_df["order_estimated_delivery_date"]
)
```
