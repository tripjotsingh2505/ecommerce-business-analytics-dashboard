```python
# OLIST_ORDERS TABLE CLEANING -->

import pandas as pd
import numpy as np
```

```python
orders = pd.read_csv("olist_orders_dataset.csv")
orders.head()
```

```python
orders.info() # To check information of the dataset including - 
# which columns are 'objects' - 
# which column have missing values - 
# Dtype is right or wrong?
```

```python
orders.shape
```

```python
date_cols = [ # Here we have taken list of strings as date_cols where we have approximately 
    # 5 attributes which we need to convert to datetime.
    "order_purchase_timestamp",
    "order_approved_at",
    "order_delivered_carrier_date",
    "order_delivered_customer_date",
    "order_estimated_delivery_date"
]

for col in date_cols:
    orders[col] = pd.to_datetime(orders[col], errors="coerce") # to_datetime() is a function in Pandas Library 
    # which is used to convert normal string(object) to datetime.
    # errors = "coerce" meaning removing error and replacing it with (NaT).
```

```python
orders.isnull().sum() # Missing value analysis.
```

```python
invalid_delivery = orders[ # Delivered date purchase se pehle to nahi?
    orders["order_delivered_customer_date"] <
    orders["order_purchase_timestamp"]
]

invalid_delivery.shape
```

```python
orders[ # Estimated date purchase se pehle to nahi?
orders["order_estimated_delivery_date"] < 
orders["order_purchase_timestamp"]
].shape
```

```python
orders["delivery_delay_days"] = ( # Raw data me ye column nahi hota
                                  # Business analysis ke liye banana padta hai.
    orders["order_delivered_customer_date"]
    - orders["order_estimated_delivery_date"]
).dt.days
```

```python
orders_delivered = orders[
     orders["order_status"] == "delivered"
].copy() # Here copy() is very important because it avoids Pandas warning.
```

```python
orders_delivered["delivery_delay_days"].describe() # Outliers sanity check.
```

```python
orders_delivered.to_csv(
    "clean_olist_orders.csv",
    index=False
) # Cleaned file bangayi.
```

```python
import pandas as pd
import numpy as np
```

```python
products = pd.read_csv("olist_products_dataset.csv")
```

```python
products.shape
```

```python
products.describe()
```

```python
products.head()
```

```python
products.isnull().sum()
```

```python
products.info()
```

```python
products.duplicated(subset="product_id").sum()
```

```python
products = products.drop_duplicates(subset="product_id")
```

```python
num_cols = [
    "product_name_lenght",
    "product_description_lenght",
    "product_photos_qty",
    "product_weight_g",
    "product_length_cm",
    "product_height_cm",
    "product_width_cm"
]

products[num_cols] = products[num_cols].apply(pd.to_numeric, errors="coerce")
```

```python
# fixing logical inconsistencies
# negative value --> NaN
# zero dimensions --> NaN.
products[num_cols] = products[num_cols].where(products[num_cols] >= 0, np.nan)
```

```python
dimension_cols = [
    "product_length_cm",
    "product_height_cm",
    "product_width_cm"
]

products[dimension_cols] = products[dimension_cols].replace(0, np.nan)
```

```python
products["product_category_name"] = products["product_category_name"].astype("string")
```

```python
products.isnull().sum()
```

```python
products.to_csv("clean_olist_products.csv", index=False)
```

```python
text_cols = ["review_comment_title", "review_comment_message"]

order_reviews[text_cols] = order_reviews[text_cols].astype("string")
```

```python
mask = order_reviews["review_answer_timestamp"] < order_reviews["review_creation_date"]
order_reviews.loc[mask, "review_answer_timestamp"] = pd.NaT
```

```python
order_reviews.isnull().sum()
```

```python
order_reviews.to_csv("clean_olist_order_reviews.csv", index=False)
```

```python
# OLIST_GEOLOCATION TABLE CLEANING -->
```

```python
import pandas as pd
import numpy as np
```

```python
geolocation = pd.read_csv("olist_geolocation_dataset.csv")
```

```python
geolocation.shape
```

```python
geolocation.describe()
```

```python
geolocation.info()
```

```python
geolocation.head()
```

```python
geolocation.duplicated().sum()
```

```python
geolocation.isnull().sum()
```

```python
geolocation = geolocation.dropna()
```

```python
geolocation["geolocation_city"] = (
    geolocation["geolocation_city"]
    .str.strip()
    .str.lower()
)

geolocation["geolocation_state"] = (
    geolocation["geolocation_state"]
    .str.strip()
    .str.upper()
)
```

```python
# taking average latitude and longitude per ZIP
geolocation_clean = (
    geolocation
    .groupby(
        ["geolocation_zip_code_prefix", "geolocation_city", "geolocation_state"],
        as_index=False
    )
    .agg({
        "geolocation_lat": "mean",
        "geolocation_lng": "mean"
    })
)
```

```python
geolocation_clean.shape
```

```python
geolocation_clean.head()
```

```python
geolocation_clean.rename(columns={
    "geolocation_zip_code_prefix": "zip_code",
    "geolocation_lat": "latitude",
    "geolocation_lng": "longitude",
    "geolocation_city": "city",
    "geolocation_state": "state"
}, inplace=True)
```

```python
geolocation_clean.to_csv(
    "clean_olist_geolocation.csv",
    index=False
)
```
